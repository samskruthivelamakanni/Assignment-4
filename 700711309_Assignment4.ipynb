```python
import numpy as np   #importing numpy
import pandas as pd  #importing pandas
import warnings    #importing warnings
warnings.filterwarnings("ignore")     #ignoring warnings

from sklearn.metrics import confusion_matrix   #importing confusion matrix
from sklearn.metrics import classification_report #importing classification report
from sklearn.model_selection import train_test_split  #importing train_test_split

# visualization
import matplotlib.pyplot as plt #importing matlab
```

```python
#question1
# 1.1 Import the given “Salary_Data.csv”

salary = pd.read_csv('Salary_Data.csv') #reading salary data set with pandas
salary
```

```python
X = salary.iloc[:, :-1].values  #considering x values
Y = salary.iloc[:, 1].values    #considering y values
```

```python
#1.2 Split the data in train_test partitions, such that 1/3 of the data is reserved as test subset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 0)
#splitting the dataset into testing & training sets
```

```python
#1.3 Train and predict the model. 
#1.4Calculate the mean_squared error

from sklearn.linear_model import LinearRegression # Importing linear regression
from sklearn.metrics import mean_squared_error    # Importing mean_squared_error
regressor = LinearRegression()           #setting regressor for LinearRegression
regressor.fit(X_train, Y_train)           #sending X & Y training datasets to the regressor
Y_pred = regressor.predict(X_test)        #Y predicting values
mean_squared_error(Y_test, Y_pred)        #Calculating mean_squared_error for Y
```

```python
#1.5 Visualize both train and test data using scatter plot.

plt.scatter(X_train,Y_train) #plot scatter for training set
plt.xlabel("Years of experience") #considering x-label as years of experience
plt.ylabel("Salary")              #considering y-label as salary
plt.show()                        #display
```

```python
plt.scatter(X_test,Y_test)           #Plot scatter test set
plt.xlabel("Years of experience")    #considering x-label as years of experience
plt.ylabel("Salary")                 #considering y-label as salary
plt.show()                           #display
```

```python
#question 2

data = pd.read_csv('K-Mean_dataset.csv') #reading K-Mean_dataset set with pandas
data #display
```

```python
#2.1 Remove any null values by the mean.

data.fillna(data.mean(),axis=0,inplace=True) #replacing nulls with mean value in  dataset
from sklearn.cluster import KMeans           #importing KMeans
x = data.drop(columns=['CUST_ID','TENURE']) #Considering X values
y = data['TENURE']                          #Considering Y values
```

```python
#2.2 Use the elbow method to find a good number of clusters with the K-Means algorithm

wcss = []                # Considering Within-Cluster Sum of Square to calculate elbow method
for i in range(1,20):      #Creating for loop for k-means range from 1 to 19
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0) # Doing K-mean
    kmeans.fit(x)          #fitting X value
    wcss.append(kmeans.inertia_) #appending iterations

plt.plot(range(1,20),wcss)   # range from 1 to 19
plt.title('the elbow method')  #title as the elbow method
plt.xlabel('Number of Clusters') #X as number of clusters
plt.ylabel('Wcss')               #y as Wcss
plt.show()                       #Display
```

```python
#2.3 Calculate the silhouette score for the above clustering

nclusters = 8 # this is the k in kmeans
km = KMeans(n_clusters=nclusters) #
km.fit(x)                    #fitting X values
y_cluster_kmeans = km.predict(x)  # calculating y values
from sklearn import metrics            #Importing metrics
score = metrics.silhouette_score(x, y_cluster_kmeans) #Calculating silhouette
print(score)                                 #display score
```

```python
#question 3
#Try feature scaling and then apply K-Means on the scaled features

from sklearn import preprocessing     #Importing preprocessing
scaler = preprocessing.StandardScaler(with_std=False) #Calculating scaler as standard deviation as False
scaler.fit(x)                        # calculating scaler values of x 
X_scaled_array = scaler.transform(x)
X_scaled = pd.DataFrame(X_scaled_array, columns = x.columns)
nclusters = 8                               # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(X_scaled)
y_cluster_kmeans = km.predict(X_scaled) #Prdicting values
from sklearn import metrics                  #Importing metrics
score = metrics.silhouette_score(X_scaled, y_cluster_kmeans)
print(score)          #display
```

```python
#Using Min-max scaler
minmax_scaler = preprocessing.MinMaxScaler()  #Calculating MinMaxScaler
minmax_scaler.fit(x)                           #fitting x values
X_scaled_array = minmax_scaler.transform(x)
X_scaled = pd.DataFrame(X_scaled_array, columns = x.columns)
nclusters = 8                         # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(X_scaled)
y_cluster_kmeans = km.predict(X_scaled)    #predicting values
from sklearn import metrics           #importing metrics
score = metrics.silhouette_score(X_scaled, y_cluster_kmeans)
print(score)                   #display
```

```python
# Using normalizer
normalizer = preprocessing.Normalizer(norm='max') #Calculating Normalizer with norm as max
normalizer.fit(x)                                  #fitting x values
X_scaled_array = normalizer.transform(x)
X_scaled = pd.DataFrame(X_scaled_array, columns = x.columns)
nclusters = 8                                        # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(X_scaled)
y_cluster_kmeans = km.predict(X_scaled)           #predicting values
from sklearn import metrics                       #importing metrics
score = metrics.silhouette_score(X_scaled, y_cluster_kmeans)
print(score)                                     #Display
```
